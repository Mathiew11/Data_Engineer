##Import Library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##Load dataset

```python
df = pd.read_csv("/content/lol_champions_updated.csv")
```

```python
df.info()
```

```python
df.head(10)
```

##Cleansing dataset

```python
df['winrate'] = df['winrate'].str.replace('%', '').astype(float)
df['banrate'] = df['banrate'].str.replace('%', '').astype(float)
df['popularity'] = df['popularity'].str.replace('%', '').astype(float)
df['tags'] = df['tags'].replace('Assasin', 'Assassin')
df['releasedate'] = pd.to_datetime(df['releasedate'], format='%d/%m/%Y')
df['release_year'] = df['releasedate'].dt.year
```

##Overview Analysis

```python
releases_per_year = df.groupby('release_year').size()

# Prepare data for plotting
years = releases_per_year.index
releases = releases_per_year.values

# Set plot style
sns.set(style="whitegrid")

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(years, releases, color="purple", marker="o", markersize=6, linewidth=1.5)

# Highlight the max and min points
max_value = releases.max()
min_value = releases.min()
max_index = releases.argmax()
min_index = releases.argmin()

# Max point
plt.scatter(years[max_index], max_value, color="green", s=100)
plt.text(years[max_index], max_value + 2, f"Max releases champion: {max_value}", color="green",ha='center', va='top')

# Min point
plt.scatter(years[min_index], min_value, color="red", s=100)
plt.text(years[min_index], min_value - 2, f"Min releases champion: {min_value}", color="red",ha='center', va='bottom')

# Labeling the axes and title
plt.title("Trend of Champion Releases Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Champions Released")

# Add a dashed vertical line for an example year (e.g., 2014)
plt.axvline(x=2013, color="gray", linestyle="--")

plt.show()
```

```python
total_hero = df['name'].count()
print("TOTAL HERO IS = " +str(total_hero))
```

```python
exploded_roles = df['role'].str.split(',').explode()

role_counts = exploded_roles.value_counts()

# Define a dictionary to map roles to specific colors
role_colors = {
    'Bottom': '#FF9999',
    'Support': '#99FF99',
    'Top': '#CC99FF',
    'Middle': '#99CCFF',
    'Jungle': '#FFCC99'

}

default_color = 'gray'
colors_assigned = [role_colors.get(role, default_color) for role in role_counts.index]
plt.figure(figsize=(12, 6))
bars = plt.bar(role_counts.index, role_counts.values, color=colors_assigned)

# Add text labels to each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, int(yval), va='bottom', ha='center')

plt.xticks(rotation=0)
plt.grid(True, which='both', linestyle='--', linewidth=0.7, color='#D3D3D3', alpha=0.8)

# Set the title and axis labels
plt.title('NUMBER OF HEROES BY ROLE')
plt.xlabel('Role')
plt.ylabel('Number of Heroes')

plt.tight_layout()
plt.show()
```

##Deep dive analysis

```python
# Set the figure size
plt.figure(figsize=(10, 6))

sns.boxplot(data=df[['winrate', 'banrate', 'popularity']], palette=["#FF9999", "#99FF99", "#9999FF"])
means = df[['winrate', 'banrate', 'popularity']].mean()

for i, mean in enumerate(means):
    plt.text(i, mean + 0.1, f'{mean:.2f}', color='black', ha='center', fontsize=12)

# Set title and labels
plt.title('Champion Winrate, Banrate, and Popularity', fontsize=14)
plt.xlabel('Stats', fontsize=12)
plt.ylabel('Value (%)', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
corr_matrix = df[['winrate', 'banrate', 'popularity', 'difficulty']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='mako', center=0, linewidths=0.5, vmin=-1, vmax=1)
plt.title('Correlation Heatmap: Winrate, Banrate, Popularity, and Difficulty')

plt.tight_layout()
plt.show()
```

```python
df['tags'] = df['tags'].str.split(',')
exploded_tags_df = df.explode('tags')
exploded_tags_df['tags'] = exploded_tags_df['tags'].replace({'Assasin': 'Assassin'})
tag_banrate_mean_df = exploded_tags_df.groupby('tags')['banrate'].mean().sort_values(ascending=False)

tag_banrate_mean_df
```

```python
plt.figure(figsize=(10, 6))

colors = sns.color_palette("pastel", 4) + ['gray'] * (len(tag_banrate_mean_df) - 4)
bars = sns.barplot(x=tag_banrate_mean_df.index, y=tag_banrate_mean_df.values, palette=colors)

mean_banrate = tag_banrate_mean_df.mean()
plt.axhline(mean_banrate, color='red', linestyle='--', linewidth=1, alpha=0.6)

plt.text(len(tag_banrate_mean_df) - 1, mean_banrate + 0.1, f'Mean = {mean_banrate:.2f}%', color='red', ha='center', fontsize=10, alpha=0.7)

# Set title and labels with enhanced formatting
plt.title('Average Banrate by Champion Tag', fontsize=14, )
plt.xlabel('Champion Tag', fontsize=12)
plt.ylabel('Average Banrate (%)', fontsize=12)

plt.xticks(rotation=45, ha='right')

# Add grid lines for better readability
plt.grid(True, which='both', linestyle='--', linewidth=0.7, color='#D3D3D3', alpha=0.8)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

tag_popularity_mean_df = exploded_tags_df.groupby('tags')['popularity'].mean().sort_values(ascending=False)

colors = sns.color_palette("pastel", 4) + ['gray'] * (len(tag_popularity_mean_df) - 4)
bars = sns.barplot(x=tag_popularity_mean_df.index, y=tag_popularity_mean_df.values, palette=colors)

mean_popularity = tag_popularity_mean_df.mean()
plt.axhline(mean_popularity, color='red', linestyle='--', linewidth=1, alpha=0.6)

plt.text(len(tag_popularity_mean_df) - 1, mean_popularity + 0.1, f'Mean = {mean_popularity:.2f}%', color='red', ha='center', fontsize=10, alpha=0.7)

# Set title and labels with enhanced formatting
plt.title('Average Popularity by Champion Tag', fontsize=14)
plt.xlabel('Champion Tag', fontsize=12)
plt.ylabel('Average Popularity (%)', fontsize=12)

plt.xticks(rotation=45, ha='right')

# Add grid lines for better readability
plt.grid(True, which='both', linestyle='--', linewidth=0.7, color='#D3D3D3', alpha=0.8)

plt.tight_layout()
plt.show()
```
